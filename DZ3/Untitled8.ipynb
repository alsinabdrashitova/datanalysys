a.	Что такое решающее дерево? Как по построенному дереву найти прогноз для
объекта?


Решающее дерево — решение задачи обучения с учителем, основанный на том, как решает задачи прогнозирования человек. 



b.	Зачем в вершинах нужны предикаты? Какие типы предикатов вы знаете? При-
ведите примеры.


Предикаты нужны для разбиения набора данных, таким образом, чтобы уменьшалось стреднее значение энтропии.
К примеру можно разбивать объекты по цвету, размеру, форме, составу и т.д.

c.	Почему для любой выборки можно построить решающее дерево, имеющее ну-
левую ошибку на ней?


На каждом листе такого дерева будет находиться один объект выборки. Такое дерево будет переобученным и не годным для работы с новыми данными.

d.	Почему не рекомендуется строить небинарные деревья (т.е. имеющие больше
двух потомков у каждой вершины)?


Это приведёт к замедлению поиска

e.	Как устроен жадный алгоритм построения дерева? Какие у него параметры?

Жадный алгоритм - на каждом шаге выбирается тот признак, при разделении по которому прирост информации оказывается наибольшим. 

f.	Зачем нужны критерии информативности?

Критерий информативности оценивает качество распределения целевой переменной среди объектов множества

g.	Как задается критерий ошибки классификации? Критерий Джини? Энтропий-
ный критерий? Какой у них смысл?


В случае критерия ошибки классификации оптимальным предсказанием будет наиболее популярный класс.
Критерий Джини является мерой, насколько часто случайно выбранный элемент из набора неверно помечается, если он случайным образом помечается согласно распределению меток в подмножестве.
Энтропийный. Воспользуемся логарифмом правдоподобия. Необходимо найтии минимум лагранжиана и дифференцировать. Подставляя полученные выражения в критерий, получим энтропию.
Это показатели, показывающие степени неравенства различных вариантов распределения доходов.
