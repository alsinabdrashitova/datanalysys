3.Верно ли, что студенты, имеющие больше 10 пропусков, учатся хуже, чем остальные?

```python
df1 = df.loc[df['absences'] > 10]
df2 = df.loc[df['absences']<= 10]
sr1 = df1['G3'].mean()
sr2 = df2['G3'].mean()
sr1, sr2
if sr1 < sr2:
    print("true")
else:
    print("false")
sr1, sr2
```

4.Как вы думаете, какие еще признаки могут повлиять на итоговую оценку? Найдите закономерности, аналогичные рассмотренным.

```python
df3 = df.loc[df['paid'] == 'yes']
df4 = df.loc[df['paid'] == 'no']
sr3 = df3['G3'].mean()
sr4 = df4['G3'].mean()
sr3, sr4
```

Кто ходит на доп. занятия учится лучше

5.Как вы думаете, если при школе откроется общежитие, насколько актуально это будет для студентов математического класса?

```python
if len(df[df['traveltime']>1]) > len(df[df['traveltime'] == 1]):
    print('True')
else:
    print('False')
```
